```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb.head(20)
```

```python
# Which hosts are the busiest and why?

host_listings = air_bnb[['host_name', 'calculated_host_listings_count']].groupby('host_name').mean().sort_values('calculated_host_listings_count', ascending=False).head(5)
host_listings

# This lists the top 5 Air BnB hosts based off of rental listings.
```

```python
# How many neighborhood groups are available and which shows up the most?

air_bnb[['neighbourhood_group', 'id']].groupby('neighbourhood_group').count()

# The below table shows the count of listings for each neighbourhood group. 
# Manhattan has the most rentals by over 1500 more than the second most popular neighbourhood.
```

```python
# Are private rooms the most popular in manhattan?

air_bnb[['room_type','neighbourhood_group', 'id']].groupby(['neighbourhood_group','room_type']).count()

# Private rooms are not the most popular in Manhattan, there are close to double the amount of entire home/apartments available.
```

```python
# Which hosts are the busiest and based on their reviews?

air_bnb[['host_name', 'reviews_per_month']].groupby(['host_name']).mean().sort_values('reviews_per_month', ascending=False).head(5)

# These are the top 5 hosts based on average reviews per month
```

```python
#Which neighorhood group has the highest average price?
air_bnb[['neighbourhood_group', 'price']].groupby('neighbourhood_group').mean().sort_values('price', ascending=False)

# Manhattan hass the most expensive rentals by average price
```

```python
# Which neighbor hood group has the highest total price?
display(air_bnb[['neighbourhood_group', 'price']].groupby('neighbourhood_group').max().sort_values('price', ascending=False))

# There is a tie between 3 of the 5 neighbourhoods for highest price for a rental


display(air_bnb[['neighbourhood_group', 'price']].groupby('neighbourhood_group').sum().sort_values('price', ascending=False))

# The second table shows the summed price over all units in each neighbourhood
```

```python
#Which top 5 hosts have the highest total price?

# Here are the top 5 hosts by total rental prices, notice Sonder(NYC) appeared on the most listings table as well

display(air_bnb[['host_name', 'price']].groupby('host_name').sum().sort_values('price', ascending=False).head(5))

# Here is a list of the top 5 most expensive rentals available, with host name, rental ID and price

display(air_bnb[['host_name', 'id', 'price']].groupby('host_name').max().sort_values('price', ascending=False).head(5))
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

# There are 17 hosts that have 0 availability with over 100 reviews. Here are their stats

air_bnb.loc[(air_bnb['availability_365'] == 0) & (air_bnb['number_of_reviews'] > 100)].head(17)
```

```python
# What host has the highest total of prices and where are they located?

# This shows the top 5 hosts by total rental prices

display(air_bnb[['host_name','host_id', "price"]].groupby('host_name').sum().sort_values('price', ascending=False).head(5))

# This shows the information for Sonder(NYC) grouped by the neighbourhood group(s) they operate in

display(air_bnb.loc[air_bnb['host_name'] == 'Sonder (NYC)'].groupby('neighbourhood_group').sum().sort_values('price',ascending=False))
```

```python
# When did Danielle from Queens last receive a review?

# This shows all 8 of the listings for Danielle in Queens. Her last review was 2019-07-08

display(air_bnb.loc[(air_bnb['host_name'] == 'Danielle') & (air_bnb['neighbourhood_group'] == 'Queens') ])
```
